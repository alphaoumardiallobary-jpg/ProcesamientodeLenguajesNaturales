# 3-Palabras vacias

```python
# importar nltk
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
# Asignar nuestras palabras vacías a una variable
en_stopwords = stopwords.words("spanish")
```

```python
# Imprima la lista de palabras vacías para ver lo que eliminaremos
print(en_stopwords)
```

```python
sentence_1 = "Era demasiado lejos para ir a la tienda y él no quería que ella caminara"
```

```python
# Mantenga las palabras en la oración si la palabra no está en la lista de palabras vacías.
sentence_no_stopwords = " ".join([word for word in sentence_1.split() if word not in (en_stopwords)])
print(sentence_no_stopwords)
```

```python
# eliminar palabras vacías de la lista
en_stopwords.remove("hemos")
en_stopwords.remove("no")
```

```python
# añadir palabras vacías personalizadas
en_stopwords.append("ir")
```

```python
sentance_no_stopwords_custom = ' '.join([word for word in sentence_1.split() if word not in (en_stopwords)])
print(sentance_no_stopwords_custom)
```

## 4.3- Sintaxis de expresiones regulares

```python
customer_reviews = ["sam me ayudó mucho en la tienda",
                    "la cajera fue muy grosera conmigo; creo que se llamaba Eleanor", 
                    "excelente trabajo de Sadeen!",
                    "sarah me ayudó a encontrar rápidamente los artículos que  necesitaba",
                    "lucy es una gran incorporación al equipo",
                    "excelente servicio de Sara, me encontró lo que quería"
                   ]
```

#### Encuentre solo las reseñas de Sarah pero tenga en cuenta la ortografía de Sara

```python
sarahs_reviews = []
pattern_to_find = r"sarah?" #El ? después de r significa que es un carácter opcional para coincidir, por lo que nuestra búsqueda buscará a Sarah y Sara.
```

```python
for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        sarahs_reviews.append(string)
```

```python
print(sarahs_reviews)
```

#### Encuentre reseñas que comiencen con la letra e

```python
e_reviews = []
pattern_to_find = r"^e" #el operador ^ para indicar el principio de una cadena 
```

```python
for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        e_reviews.append(string)
```

```python
print(e_reviews)
```

#### Encuentre reseñas que comiencen con la letra a

```python
a_reviews = []
pattern_to_find = r"a$" #el operador $ para indicar el final de una cadena 
```

```python
for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        a_reviews.append(string)
```

```python
print(a_reviews)
```

#### Encuentre reseñas que contengan las palabras necesarias o deseadas

```python
needwant_reviews = []
pattern_to_find = r"(necesitaba|quería)" # el operador de tubería | puede usarse para significar O
```

```python
for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        needwant_reviews.append(string)
```

```python
print(needwant_reviews)
```

#### Eliminar cualquier cosa de la reseña que no sea una palabra o un espacio (es decir, eliminar la puntuación)

```python
no_punct_reviews = []
pattern_to_find = r"[^\w\s]" 
# [^ ] significa "no", \w significa palabra y \s significa espacio en blanco: así que busque cualquier cosa que no sea una palabra o un espacio.
```

```python
for string in customer_reviews:
    no_punct_string = re.sub(pattern_to_find, "", string)
    no_punct_reviews.append(no_punct_string)
```

```python
print(no_punct_reviews)
```

# 8- Abreviación

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['El','surgimiento', 'de', 'la', 'inteligencia', 'artificial', 'ha', 'llevado', 'a', 'avances', 'significativos', 'en', 'el', 'procesamiento', 'del', 'lenguaje', 'natural', 'la', 'visión', 'y', 'otros', 'campos', 'del', 'aprendizaje', 'automático', 'así', 'como', 'en', 'los', 'algoritmos', 'permite', 'que', 'las', 'computadoras', 'realicen', 'tareas', 'complejas', 'que', 'se', 'creía', 'que', 'eran', 'exclusivas', 'de', 'los', 'humanos', 'han', 'convertido', 'incluso', 'más', 'poderoso', 'capaz', 'de', 'procesar', 'vastas', 'cantidades', 'de', 'datos', 'y', 'aprender', 'de', 'ello', 'en', 'formas', 'que', 'no', 'eran', 'posibles', 'antes', 'como', 'un', 'resultado', 'ia', 'es', 'cada', 'vez', 'más', 'siendo', 'usado', 'en', 'una', 'amplia', 'gama', 'de', 'industrias', 'desde', 'atención', 'médica', 'hasta', 'finanzas', 'hasta', 'transporte', 'y', 'su', 'impacto', 'es', 'solo', 'establecido', 'para', 'crecer', 'en', 'los', 'años', 'hasta', 'venir'
]
print(tokens)
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())
print(unigrams)
```

```python
unigrams[0:10].sort_values().plot.barh(color='lightsalmon', width=.9, figsize=(12, 8))
plt.title('Los 10 unigramas más frecuentes')
plt.ylabel('Unigram')
plt.xlabel(' De ocurrencias')
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(bigrams)
```

```python
bigrams[0:10].sort_values().plot.barh(color='skyblue', width=.9, figsize=(12, 8))
plt.title('Los 10 bigramas más frecuentes')
plt.ylabel('Bigram')
plt.xlabel('De ocurrencias')
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts()) 
print(trigrams)
```
