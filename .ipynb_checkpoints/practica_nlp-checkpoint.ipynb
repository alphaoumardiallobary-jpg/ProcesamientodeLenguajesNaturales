# Practica de nlp

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

### Cargando Datos

```python
data = pd.read_csv("tripadvisor_hotel_reviews (1).csv", sep=";", engine="python")
```

```python
data.info()
```

```python
data.head(5)
```

```python
data["Revisar"][0]
```

# Minusculas

```python
data["Revisar_lowercase"]= data["Revisar"].str.lower()
```

```python
data.head(5)
```

# Eliminación de palabras vacías

```python
en_stopwords = stopwords.words("spanish")
en_stopwords.remove("no")
```

```python
data["Revisar_no_stopwords"] = data["Revisar_lowercase"] \
                                .apply(lambda x: ' ' \
                                .join([word for word \
                                in x.split() if word \
                                not in (en_stopwords)]))
```

```python
data.head(5)
```

```python
data["Revisar_no_stopwords"][0]
```

# Puntuación

```python
data["Revisar_no_stopwords_no_punct"] = data \
                                    .apply(lambda x: \
                                    re.sub(r"[*]", \
                                    "star", \
                                    x["Revisar_no_stopwords"] \
                                    ), axis=1)
```

```python
data.head(5)
```

```python
data["Revisar_no_stopwords_no_punct"] = data. \
                                    apply(lambda x: \
                                    re.sub(r"([^\w\s])", \
                                    "", \
                                    x["Revisar_no_stopwords_no_punct"] \
                                    ), axis=1)
```

```python
data.head(5)
```

```python
data["Revisar_no_stopwords_no_punct"][0] # observamos que todo esta en minuscula, no hay puntuaciones y tampoco simbolos
```

# Tokenización

```python
data["tokenized"] = data.apply(lambda x: \
                               word_tokenize( \
                               x["Revisar_no_stopwords_no_punct"] \
                               ), axis=1)
```

```python
data.head(5)
```

```python
data["tokenized"][0]
```

# Lematización

```python
lemmatizer = WordNetLemmatizer()
```

```python

data["lemmatized"] = data["tokenized"] \
                    .apply(lambda tokens: \
                    [lemmatizer.lemmatize(token) \
                     for token in tokens])
```

```python
data.head(5)
```

```python
data["lemmatized"][0]
```

# Derivación

```python
ps = PorterStemmer()
```

```python
data["stemmed"] = data["tokenized"] \
                  .apply(lambda tokens: \
                  [ps.stem(token) \
                   for token in tokens])
```

```python
data.head(5)
```

```python
data["stemmed"][0]
```

# Abreviación (unigrams)

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
# unigrams: n=1
unigrams = (pd.Series \
            (nltk.ngrams(tokens_clean, 1)) \
            .value_counts()) 
print(unigrams)
```

```python
# bigrams: n=2
bigrams = (pd.Series \
           (nltk.ngrams(tokens_clean, 2)) \
           .value_counts()) 
print(bigrams)
```

```python
ngrams_4 = (pd.Series \
            (nltk.ngrams(tokens_clean, 4)) \
            .value_counts()) 
print(ngrams_4)
```
